## 2. IMPORT Library

```
import numpy as np
import pandas as pd
import pickle
import tensorflow as tf
from tensorflow import keras
print(f"TF: {tf.__version__}")
```

## 3. DEFINISI CUSTOM LOSS (WAJIB ADA SEBELUM LOAD MODEL)

```
class FocalLoss(keras.losses.Loss):
    def __init__(self, gamma=2.0, alpha=0.25, name='focal_loss', **kwargs):
        super().__init__(name=name, **kwargs)
        self.gamma = gamma
        self.alpha = alpha

    def call(self, y_true, y_pred):
        y_true    = tf.cast(tf.reshape(y_true, [-1]), tf.int32)
        y_true_oh = tf.cast(tf.one_hot(y_true, depth=3), tf.float32)
        y_pred    = tf.cast(y_pred, tf.float32)
        y_pred    = tf.clip_by_value(y_pred, 1e-7, 1.0 - 1e-7)
        ce           = -y_true_oh * tf.math.log(y_pred)
        p_t          = tf.reduce_sum(y_true_oh * y_pred, axis=-1, keepdims=True)
        focal_weight = tf.pow(1.0 - p_t, self.gamma)
        loss         = self.alpha * focal_weight * tf.reduce_sum(ce, axis=-1)
        return tf.reduce_mean(loss)

    def get_config(self):
        cfg = super().get_config()
        cfg.update({'gamma': self.gamma, 'alpha': self.alpha})
        return cfg

print("FocalLoss OK.")
```

## 4. LOAD MODEL & ARTIFACTS

```
model = keras.models.load_model(
    path_model + 'worksafe_model_v1.keras',
    custom_objects={'FocalLoss': FocalLoss}
)

with open(path_model + 'scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

with open(path_model + 'imputer.pkl', 'rb') as f:
    imputer = pickle.load(f)

with open(path_model + 'feature_cols.pkl', 'rb') as f:
    feature_cols = pickle.load(f)

LABEL_MAP = {0: 'Low Risk', 1: 'Medium Risk', 2: 'High Risk'}

# Simpan nilai median dari imputer untuk default input
median_values = {
    col: float(imputer.statistics_[i])
    for i, col in enumerate(feature_cols)
}

print("Model & artifacts loaded.")
print(f"Jumlah fitur: {len(feature_cols)}")
```

##  5. FUNGSI PREPROCESSING INPUT

```
def preprocess_input(input_dict):
    df = pd.DataFrame([input_dict])

    for col in feature_cols:
        if col not in df.columns:
            df[col] = np.nan

    df = df[feature_cols]

    df_imp = pd.DataFrame(
        imputer.transform(df),
        columns=feature_cols
    )

    df_scaled = pd.DataFrame(
        scaler.transform(df_imp),
        columns=feature_cols
    )

    return df_scaled.values.astype(np.float32)

print("preprocess_input OK.")
```

## 6. FUNGSI PREDIKSI

```
def predict_risk(input_dict):
    X     = preprocess_input(input_dict)
    proba = model.predict(X, verbose=0)[0]

    kelas = int(np.argmax(proba))
    label = LABEL_MAP[kelas]

    return {
        'risk_label'   : label,
        'risk_class'   : kelas,
        'confidence'   : round(float(proba[kelas]) * 100, 2),
        'probabilities': {
            'Low Risk'   : round(float(proba[0]) * 100, 2),
            'Medium Risk': round(float(proba[1]) * 100, 2),
            'High Risk'  : round(float(proba[2]) * 100, 2),
        }
    }
```

## 7. CONTOH INFERENSI

```
print("=" * 50)
print("  WorkSafe AI - Cek Risiko Pekerjaan Kamu")
print("=" * 50)
print("Jawab setiap pertanyaan dengan angka 1 sampai 5.")
print("1 = Sangat Rendah  |  3 = Sedang  |  5 = Sangat Tinggi")
print("-" * 50)

def tanya(pertanyaan, nama_fitur):
    while True:
        try:
            val = float(input(f"{pertanyaan} (1-5): "))
            if 1.0 <= val <= 5.0:
                return nama_fitur, val
            print("  Angka harus antara 1 sampai 5.")
        except ValueError:
            print("  Masukkan angka saja.")

pertanyaan_list = [
    ("Seberapa sering kamu benerin atau rawat mesin di kerjaan?",          "act_repairing_and_maintaining_mechanical_equipment"),
    ("Seberapa sering kamu operasiin mesin atau kendaraan?",               "act_operating_vehicles,_mechanized_devices,_or_equipment"),
    ("Seberapa sering kamu angkat atau pindahin barang fisik?",            "act_handling_and_moving_objects"),
    ("Kalau ada yang error, seberapa jago kamu nyari masalahnya?",         "skl_troubleshooting"),
    ("Seberapa sering kerjaan kamu butuh tenaga fisik besar?",             "act_performing_general_physical_activities"),
    ("Seberapa sering kamu mikir dalam-dalam sebelum ambil keputusan?",    "skl_critical_thinking"),
    ("Seberapa sering kamu hadapin masalah rumit yang butuh solusi kreatif?", "skl_complex_problem_solving"),
    ("Seberapa sering kamu olah atau analisis data di kerjaan?",           "act_analyzing_data_or_information"),
    ("Seberapa sering kamu ambil keputusan penting?",                      "skl_judgment_and_decision_making"),
    ("Seberapa mudah kamu nyerap hal baru tanpa harus diajarkan terus?",   "skl_active_learning"),
    ("Seberapa sering kamu ngomong atau presentasi di depan orang?",       "skl_speaking"),
    ("Seberapa sering kamu nulis laporan atau dokumen?",                   "skl_writing"),
    ("Seberapa sering kamu negosiasi atau cari jalan tengah sama orang lain?", "skl_negotiation"),
    ("Seberapa besar porsi kerjaanmu yang pakai komputer?",                "act_working_with_computers"),
    ("Seberapa sering kamu bikin rencana dan atur prioritas kerja?",       "act_organizing,_planning,_and_prioritizing_work"),
    ("Seberapa sering kamu koordinasi sama tim?",                          "skl_coordination"),
    ("Seberapa sering orang minta saran profesional dari kamu?",           "act_providing_consultation_and_advice_to_others"),
    ("Seberapa baik kamu ngatur waktu biar semua selesai tepat waktu?",    "skl_time_management"),
    ("Seberapa sering kamu pantau jalannya proses atau kondisi sekitar?",  "act_monitoring_processes,_materials,_or_surroundings"),
    ("Seberapa rajin kamu update ilmu yang relevan sama kerjaanmu?",       "act_updating_and_using_relevant_knowledge"),
]

sample_input = {}
for pertanyaan, fitur in pertanyaan_list:
    fitur_key, nilai = tanya(pertanyaan, fitur)
    sample_input[fitur_key] = nilai

# 30 fitur sisanya diisi median dari data training otomatis
for col in feature_cols:
    if col not in sample_input:
        sample_input[col] = median_values[col]

print("\nMemproses...")
result = predict_risk(sample_input)

print("\n" + "=" * 50)
print("  HASIL ANALISIS RISIKO PEKERJAAN KAMU")
print("=" * 50)
print(f"  Kategori Risiko : {result['risk_label']}")
print(f"  Confidence      : {result['confidence']}%")
print("-" * 50)
print("  Probabilitas tiap kategori:")
for lbl, prob in result['probabilities'].items():
    bar = '█' * int(prob / 5)
    print(f"  {lbl:<14}: {prob:>6}%  {bar}")
print("=" * 50)

if result['risk_class'] == 0:
    print("\n  Pekerjaan kamu tergolong AMAN dari risiko otomasi.")
    print("  Skill kamu didominasi kemampuan kognitif yang")
    print("  sulit digantikan mesin.")
elif result['risk_class'] == 1:
    print("\n  Pekerjaan kamu punya risiko SEDANG.")
    print("  Sebagian tugasmu berpotensi terotomasi, tapi")
    print("  masih banyak yang butuh sentuhan manusia.")
else:
    print("\n  Pekerjaan kamu punya risiko TINGGI terkena otomasi.")
    print("  Mulai kembangkan skill digital dan kognitif")
    print("  biar tetap relevan di dunia kerja.")
```

## 8. BATCH INFERENCE (OPSIONAL)

```
df_infer = pd.read_csv(path_processed + 'test_data.csv').head(10)

# data test_data.csv sudah di-scale, langsung masuk model
results = []
for _, row in df_infer.iterrows():
    X     = row[feature_cols].values.astype(np.float32).reshape(1, -1)
    proba = model.predict(X, verbose=0)[0]
    kelas = int(np.argmax(proba))
    results.append({
        'actual_label': LABEL_MAP[int(row['risk_label'])],
        'predicted'   : LABEL_MAP[kelas],
        'confidence'  : round(float(proba[kelas]) * 100, 2),
        'match'       : '✓' if LABEL_MAP[int(row['risk_label'])] == LABEL_MAP[kelas] else '✗'
    })

df_results = pd.DataFrame(results)
print("Hasil batch inference (10 sampel):")
display(df_results)

correct = df_results['match'].value_counts().get('✓', 0)
print(f"\nAkurasi di 10 sampel ini: {correct}/10")
```
